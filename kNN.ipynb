### 4. Split the data into training set and test set. Make a test set size 0.3 and random seed 123 (so the results are consistent for all students)

```python
y = df['TARGET CLASS']
x = df.drop(columns=['TARGET CLASS'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=123)
```

### 5. Standardize the variables - scale them to zero mean and unit variance. Do not pass the "TARGET CLASS" column to the scaler!

```python
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

### 6. Check that the features are indeed scaled

```python
plt.plot(x, y)
```

```python
plt.plot(X_train, Y_train)
```

### 7. Fit the kNN to the training data. Use k = 1.

```python
n_neighbors = 1

clf = neighbors.KNeighborsClassifier(n_neighbors)
clf.fit(X_train, Y_train)
```

### 8. Check the classifier performance

```python
pred = clf.predict(X_test)
```

```python
print(confusion_matrix(Y_test, pred))
```

```python
print(classification_report(Y_test, pred))
```

### 9. Try different k values and select one that works best. The way to do it is to see when increasing k does not add to performance anymore.

```python
n_neighbors = 7

clf = neighbors.KNeighborsClassifier(n_neighbors)
clf.fit(X_train, Y_train)

pred = clf.predict(X_test)
print(confusion_matrix(Y_test, pred))
```

```python
print(classification_report(Y_test, pred))
```

```python
# k = 7 is best
# you want the precision cloes to 1 but confusion to have highest numbers in diagonal and smallest numbers everywhere else
```
